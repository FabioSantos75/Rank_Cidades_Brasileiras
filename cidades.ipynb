**ANÁLISE DO ESTADO DO PARÁ**

```python
para = df[df['SIGLA_UF'] == 'PA']
qtd_para = df[df['SIGLA_UF'] == 'PA'].shape[0]

maior_para = para['IDSC-BR_2024'].max()
media_pais = round(df['IDSC-BR_2024'].mean(), 2)
media_para = para['IDSC-BR_2024'].mean()

piores_para = para[para['IDSC-BR_2024'] < media_pais].shape[0]
melhores_para = para[para['IDSC-BR_2024'] > media_pais].shape[0]

maiores_para = para.loc[para['IDSC-BR_2024'] > media_pais]

print(f'Média do Brasil - {media_pais}')
print(f'Média Pará - {round(media_para, 2)}')
print(f'Maior Pará - {maior_para}')

print()
print('Quantidade de Cidades do Pará abaixo da Média do Brasil:')
print(f'{piores_para} -- {round(piores_para * 100 / qtd_para, 2)}%')
print('-'*30)
print('Acima da Média:')
print(f'{melhores_para} -- {round(melhores_para * 100 / qtd_para, 2)}%')
print(maiores_para)
```

Nível de Desenvolvimento Sustentável:

Muito alto - 80 a 100

Alto - 60 a 79,99

Médio - 50 a 59,99

Baixo - 40 a 49,99

Muito baixo - 0 a 39,99

```python
media = round(df['IDSC-BR_2024'].mean(), 2)

print(f'Média das Cidades Brasileiras - {media}')
print('RESULTADO:')
if 80 <= media <= 100:
    print('Muinto alto')

elif 60 <= media <= 79.99:
    print('Alto')

elif 50 <= media <= 59.99:
    print('Médio')

elif 40 <= media <= 49.99:
    print('Baixo')

elif 0 <= media <= 39.99:
    print('Muito baixo')
else:
    print('Informáções Indisponíveis')

```

```python
municipios = df['COD_MUN'].nunique()
pont = df['IDSC-BR_2024']

muito_alto = df[(pont > 80) & (pont <= 100)]['MUNICIPIO'].count()

alto = df[(pont >= 60) & (pont <= 79.99)]['MUNICIPIO'].count()

medio = df[(pont>= 50) & (pont <= 59.99)]['MUNICIPIO'].count()

baixo = df[(pont >= 40) & (pont <= 49.99)]['MUNICIPIO'].count()

muito_baixo = df[(pont > 0) & (pont <= 39.99)]['MUNICIPIO'].count()

print(f'Municipios - {municipios}')
print()
print(f'Muito Alto - {muito_alto} -- {round(muito_alto * 100 / municipios, 2)} %')

print(f'Alto - {alto} -- {round(alto * 100 / municipios, 2)} %')

print(f'Médio - {medio} -- {round(medio * 100 / municipios, 2)} %')

print(f'Baixo - {baixo} -- {round(baixo * 100 / municipios, 2)} %')

print(f'Muito Baixo - {muito_baixo} -- {round(muito_baixo * 100 / municipios, 2)} %')


```
